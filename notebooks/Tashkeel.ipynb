# Partial Arabic Vocalization

This notebook derives the partially vocalized Arabic representation from Mishkal output. The implementation is in `scripts/create_partial_vocalization.py`.

The transformation preserves internal diacritics and terminal shadda while removing word-final short vowels, tanween, and sukun. It also handles fathatan encoded before a supporting final alif.

```python
from pathlib import Path
import importlib.util
import sys

REPO_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
SCRIPT_PATH = REPO_ROOT / "scripts" / "create_partial_vocalization.py"
spec = importlib.util.spec_from_file_location("partial_vocalization", SCRIPT_PATH)
module = importlib.util.module_from_spec(spec)
sys.modules[spec.name] = module
spec.loader.exec_module(module)
remove_word_final_endings = module.remove_word_final_endings
```

```python
examples = [
    "الطَّالِبُ مُجْتَهِدٌ",
    "مَرْحَبًا، بِكُمْ!",
    "مَرّ",
]
[(text, remove_word_final_endings(text)) for text in examples]
```

Expected behavior:

- `الطَّالِبُ مُجْتَهِدٌ` → `الطَّالِب مُجْتَهِد`
- internal marks such as fatha, kasra, sukun, and shadda remain unchanged
- final punctuation remains unchanged
- a terminal shadda is not removed

```python
import subprocess

INPUT_PATH = REPO_ROOT / "data" / "processed" / "vocalized_demo.tsv"
OUTPUT_PATH = REPO_ROOT / "data" / "processed" / "final_parallel_demo.tsv"
LIMIT = 100

command = [
    sys.executable,
    str(SCRIPT_PATH),
    "--input", str(INPUT_PATH),
    "--output", str(OUTPUT_PATH),
    "--overwrite",
]
if LIMIT is not None:
    command.extend(["--limit", str(LIMIT)])

print(" ".join(command))
# Uncomment to execute:
# subprocess.run(command, check=True)
```
